```python
'''SKD
create meta file'''
data_folder = '/bigdata/siyiplace/data/skin_lesion'
melanoma_folder = data_folder+'/skin_cancer_detection_origin/skin_image_data_set-1/Skin Image Data Set-1/skin_data/melanoma'
benign_folder = data_folder+'/skin_cancer_detection_origin/skin_image_data_set-2/Skin Image Data Set-2/skin_data/notmelanoma'
image_name_list = []
diagnosis_list = []

for dataset_name in ['dermis', 'dermquest']:
    path_list = os.listdir('{}/{}'.format(melanoma_folder, dataset_name))
    for path in path_list:
        if path[-4:] == '.jpg':
            image_name = dataset_name+'_'+path[:-4]
            image_name_list.append(image_name)

diagnosis_list.extend(['Melanoma']*len(image_name_list))

count = 0
for dataset_name in ['dermis', 'dermquest']:
    path_list = os.listdir('{}/{}'.format(benign_folder, dataset_name))

    for path in path_list:
        if path[-4:] == '.jpg':
            image_name = dataset_name+'_'+path[:-4]
            image_name_list.append(image_name)
            count += 1

diagnosis_list.extend(['Benign']*count)
```

```python
print(len(diagnosis_list))
print(len(image_name_list))
SKD_meta_path = data_folder+'/SKD/meta_SKD.csv'
df = pd.DataFrame({
    'dataset': ['SKD']*len(diagnosis_list),
    'ID': image_name_list,
    'diagnosis': diagnosis_list,
})
df.to_csv(SKD_meta_path, index=False)
```
